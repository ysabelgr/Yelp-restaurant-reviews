```python
reviews1 = reviews.dropna(how='any',axis=0)
```

```python
#check number of null values

reviews1.isnull().sum()
```

```python
# adding new column state
state = []
for row in reviews1.location.index:
    state.append(reviews1.location[row].split(' ')[-1])
```

```python
reviews1['state']=state
```

```python
#Word count
```

```python
word_count = []
for row in reviews1.content.index:
    word_count.append(len([i for i in reviews1.content[row].split(' ')]))
```

```python
reviews1['word_count']=word_count
```

```python
reviews1
```

```python
# Density of word count for each star rating

sns.set(style='darkgrid',font_scale=2)
snsplt = plt.figure(figsize=(8,8))
snsplt = plt.title('Number of Words in reviews')
snsplt=sns.kdeplot(reviews1[reviews1.stars == 1]['word_count'],shade=True, color="#e32019", label='1 star')
snsplt=sns.kdeplot(reviews1[reviews1.stars == 2]['word_count'], shade=True, color="#f2870c",label='2 stars')
snsplt=sns.kdeplot(reviews1[reviews1.stars == 3]['word_count'], shade=True, color="#f2eb0c",label='3 stars')
snsplt=sns.kdeplot(reviews1[reviews1.stars == 4]['word_count'], shade=True, color="#0cf2eb",label='4 stars')
snsplt=sns.kdeplot(reviews1[reviews1.stars == 5]['word_count'], shade=True, color="#0cf249",label='5 stars')

p=snsplt.get_figure()
p.savefig('/Users/ysabelgonzalezrico/Documents/UCI/4 - SPRING/NLP/Yelp/wordcnt.png',transparent=True)
```
